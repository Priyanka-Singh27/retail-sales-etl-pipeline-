```python
df['Item_Weight'] = df['Item_Weight'].fillna(df.groupby('Item_Type')['Item_Weight'].transform('mean'))
```

```python

df['Outlet_Size'] = df['Outlet_Size'].fillna(
    df.groupby('Outlet_Type')['Outlet_Size'].transform(lambda x: x.mode()[0] if not x.mode().empty else np.nan)
)
```

```python
fat_content_map = {
    'low fat': 'Low Fat',
    'LF': 'Low Fat',
    'reg': 'Regular',
    'regular': 'Regular'
}
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace(fat_content_map)
```

```python
import numpy as np
df['Item_Visibility'] = df['Item_Visibility'].replace(0, np.nan)
df['Item_Visibility'] = df['Item_Visibility'].fillna(df.groupby(['Item_Type','Outlet_Type'])['Item_Visibility'].transform('mean'))
```

```python
df.to_csv('cleaned_bigmart_data.csv', index=False)
```
